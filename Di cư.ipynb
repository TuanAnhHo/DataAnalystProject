```python
### drop some unnecessary columns and rename some columns:
df2 = pd.read_stata('/Users/hoanh/Desktop/Data Source/10 - VHLSS 2018/2 - Data/1 - Households/MUC1B.dta')
df_muc1b = df2.drop(columns=['m1bc10','m1bc11n','m1bc11t'])
df_muc1b = df_muc1b.rename(columns={'m1bc3':'gioitinh','m1bc5':'age','m1bc6':'relationship','m1bc7a':'GDPT&CĐ','m1bc7b':'GDNN','m1bc2':'name'})
```

```python
### Không có bằng cấp 0
### tiểu học 1
### thcs 2
### thpt 3
### sơ cấp nghề 4
### trung cấp nghề 5
### trung học chuyên nghiệp 6
### cao đẳng nghề 7
### cao đẳng 8
### đại học 9
### thạc sĩ 10
### tiến sĩ 11
### khác 12
```

```python
### covert education of migrants into dummy variables
### create new columns which called 'edu' 

edu1, edu2, edu = [], [], []

for i in df_muc1b['GDPT&CĐ']:
    if i == 'Tiểu học': edu1.append(1)
    if i == 'THPT': edu1.append(3)
    if i == 'THCS': edu1.append(2)
    if i == 'Đại học': edu1.append(9)
    if i == 'K0 bằng cấp': edu1.append(0)
    if i == 'Cao đẳng': edu1.append(8)
    if i == 'Khác': edu1.append(12)
    if i == 'Thạc sĩ': edu1.append(10)
    if i == 'Tiến sĩ': edu1.append(11)

for z in df_muc1b['GDNN']:
    if z == 'Không': edu2.append(0)
    if z == 'Sơ cấp nghề': edu2.append(4)
    if z == 'TH CN': edu2.append(6)
    if z == 'Cao đẳng nghề': edu2.append(7)
    if z == 'Trung cấp nghề': edu2.append(5)
    
for t in range(len(edu1)):
    if edu1[t] > edu2[t]: edu.append(edu1[t])
    else: edu.append(edu2[t])

df_muc1b['edu'] = edu
```

```python
# convert sex variable to dummy variable
sex = []
for i in df_muc1b['gioitinh']:
    if i == 'Nữ': sex.append(0)
    if i == 'Nam': sex.append(1)
df_muc1b['sex'] = sex
```

```python
### extract some essential columns for finding migrants
df_muc1b = df_muc1b[['tinh','huyen','xa','diaban','hoso','name','age','sex','edu','m1bc4']]
```
