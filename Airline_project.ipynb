```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### SUMMARY THE NUMBER OF PASSENGERS TO AND FROM INDIA BETWEEN 2015 AND Q1,2017

```python
graph1 = pd.read_csv('/Users/hoanh/Desktop/Portfolio/Airline Project/Data for Python/graph1.csv')
graph1['QUARTER-YEAR'] = graph1['YEAR'].map(str) + '-' + graph1['QUARTER'].map(str)
fig, axes = plt.subplots(1,2,figsize=(10,3.5))
fig.suptitle('The number of passengers to and from INDIA', y=1.1)

sns.lineplot(ax=axes[0],data=graph1, x='QUARTER-YEAR', y='PASSENGERS TO INDIA', color='r')
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=45, fontsize=9)
axes[0].set_title('TO INDIA', fontsize= 11)
axes[0].set_xlabel('')
axes[0].set_ylabel('The number of passenger')
plt.style.use('_mpl-gallery-nogrid')
sns.lineplot(ax=axes[1],data=graph1, x='QUARTER-YEAR', y='PASSENGERS FROM INDIA')
axes[1].set_xticklabels(axes[0].get_xticklabels(),rotation=45,fontsize=9)
axes[1].set_title('FROM INDIA', fontsize= 11)
axes[1].set_xlabel('')
axes[1].set_ylabel('The number of passenger')

plt.show()
```
